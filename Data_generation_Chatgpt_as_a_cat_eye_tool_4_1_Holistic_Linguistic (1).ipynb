```python
import os
import re
from openai import OpenAI

# === CONFIGURATION ===
API_KEY = "9b22bd964687d47bee866413d046c5a5"
BASE_URL = "https://chat-ai.academiccloud.de/v1"
MODEL = "qwen2.5-vl-72b-instruct"
FOLDER = "Original"

# === INITIALIZE CLIENT ===
client = OpenAI(api_key=API_KEY, base_url=BASE_URL)

# === DEFINE PROMPTS ===

def prompt_easy(text):
    return f"""übersetzen Sie den folgenden Text in die Deutsche Leichte Sprache:

\"\"\"
{text}
\"\"\"
"""

def prompt_easier(text):
    return f"""linguistischer Ansatz (Sprachebenen): reformuliere den Text, aber:

- ohne unwichtige Informationen (Textebene)
- ohne Nebensätze und mit einfachen syntaktischen Strukturen (Satzebene)
- füge Erklärungen für schwierige Wörter hinzu (Wortebene)

Text:
\"\"\"
{text}
\"\"\"
"""

# === PROCESS EACH FILE ===
for filename in os.listdir(FOLDER):
    match = re.match(r"(BN|K)_(\d+)B\.txt", filename)
    if not match:
        continue

    prefix = match.group(1)  # BN or K
    number = match.group(2)
    base_path = os.path.join(FOLDER, filename)

    # Load original text
    with open(base_path, "r", encoding="utf-8") as f:
        original_text = f.read()

    # === STEP 1: Leichte Sprache ===
    messages_easy = [
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": prompt_easy(original_text)}
    ]

    response_easy = client.chat.completions.create(
        model=MODEL,
        messages=messages_easy
    )

    easy_text = response_easy.choices[0].message.content.strip()

    # Save Leichte Sprache version
    easy_filename = f"{prefix}_{number}E.txt"
    with open(os.path.join(FOLDER, easy_filename), "w", encoding="utf-8") as f:
        f.write(easy_text)
    print(f"Saved Leichte Sprache: {easy_filename}")

    # === STEP 2: Noch einfacher mit linguistischem Ansatz ===
    messages_easier = [
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": prompt_easier(easy_text)}
    ]

    response_easier = client.chat.completions.create(
        model=MODEL,
        messages=messages_easier
    )

    easier_text = response_easier.choices[0].message.content.strip()

    # Save Noch Einfacher version
    easier_filename = f"{prefix}_{number}H.txt"
    with open(os.path.join(FOLDER, easier_filename), "w", encoding="utf-8") as f:
        f.write(easier_text)
    print(f"Saved Noch Einfacher: {easier_filename}")
```
